Write a code to replace all the missing values with mean for numerical attributes and mode for categorical attributes and 
print the dataframe.

﻿

```python
# calling Libraries

import pandas as pd
import numpy as np

# creating the dataframe

data = [[np.NaN,42,'First Batting',np.NaN],[48,39,'First Batting','not out'],[75,51,'Chasing','out'],[15,8,'Chasing','out']
        ,[32,27,'First Batting','not out'],[21,np.NaN,'First Batting','out'],[np.NaN,26,'Chasing','not out'],[9,6,'Chasing','out'],
       [32,np.NaN,'First Batting','out'],[np.NaN,19,'First Batting',np.NaN],[52,23,'Chasing',np.NaN],[np.NaN,6,'Chasing','not out']]
df = pd.DataFrame(data,columns=['Runs_Scored','Balls_Faced','First_Batting/Chasing','Wicket'])

# Write a code to replace all the missing values with mean for numerical attributes and mode for categorical attributes and print the dataframe.

df['Runs_Scored'].fillna(df['Runs_Scored'].mean(),inplace = True)
df['Balls_Faced'].fillna(df['Balls_Faced'].mean(),inplace = True)

df['First_Batting/Chasing'].fillna(df['First_Batting/Chasing'].mode()[0],inplace = True)
df['Wicket'].fillna(df['Wicket'].mode()[0],inplace = True)

print(df)
```

Write a code to print the dispersion measures for the attribute 'Runs_Scored' from the given dataframe.
﻿

1- print for minimum runs scored

2- print for maximum runs scored

3- print for range of runs scored

4- print for 25th percentile of runs scored

5- print for 50th percentile of runs scored

6- print for 75th percentile of runs scored

7- print for variance of runs scored

8- print for standard deviation of runs scored

﻿





Sample Output :


```python
# calling Libraries

import pandas as pd
import numpy as np

# creating the dataframe

data = [[np.NaN,42,'First Batting',np.NaN],[48,39,'First Batting','not out'],[75,51,'Chasing','out'],[15,8,'Chasing','out']
        ,[32,27,'First Batting','not out'],[21,np.NaN,'First Batting','out'],[np.NaN,26,'Chasing','not out'],[9,6,'Chasing','out'],
       [32,np.NaN,'First Batting','out'],[np.NaN,19,'First Batting',np.NaN],[52,23,'Chasing',np.NaN],[np.NaN,6,'Chasing','not out']]
df = pd.DataFrame(data,columns=['Runs_Scored','Balls_Faced','First_Batting/Chasing','Wicket'])

# Write a code to print the dispersion measures for the attribute 'Runs_Scored' from the given dataframe.

# 1- print for minimum runs scored
print(df["Runs_Scored"].min())

# 2- print for maximum runs scored
print(df["Runs_Scored"].max())

# 3- print for range of runs scored
print(df["Runs_Scored"].max()-df["Runs_Scored"].min())

# 4- print for 25th percentile of runs scored
print(df["Runs_Scored"].quantile(0.25))

# 5- print for 50th percentile of runs scored
print(df["Runs_Scored"].quantile(0.5))

# 6- print for 75th percentile of runs scored
print(df["Runs_Scored"].quantile(0.75))

# 7- print for variance of runs scored
print(df["Runs_Scored"].var())

# 8- print for standard deviation of runs scored
print(df["Runs_Scored"].std())
```
